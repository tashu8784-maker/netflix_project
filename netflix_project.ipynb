```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('netflix_titles_csv.zip')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#MISSING VALUE TREATMENT-
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['director'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
# Extract year & month from date_added
df['year_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year
df['month_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.month
```

```python
#EXPLORATORY DATA ANALYSIS-
type_count=df['type'].value_counts()
type_count
```

```python
#MOVIES VS TV SHOWS-
sns.countplot(data=df, x='type', palette="Set2")
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# Top 10 countries with most Netflix content
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="muted")
plt.title("Top 10 Countries with Most Content")
plt.xticks(rotation=45)
plt.show()
```

```python
# Content added per year
df['year_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year
content_per_year = df['year_added'].value_counts().sort_index()
plt.figure(figsize=(10,5))
content_per_year.plot(kind='line', marker='o')
plt.title("Netflix Content Added Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Top 10 genres-
from collections import Counter
genres = df['listed_in'].dropna().apply(lambda x: x.split(', '))
genre_list = [g for sublist in genres for g in sublist]
top_genres = Counter(genre_list).most_common(10)
genre_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])
genre_df
```

```python
#visualization of Top 10 genres-
from collections import Counter
genres = df['listed_in'].dropna().apply(lambda x: x.split(', '))
genre_list = [g for sublist in genres for g in sublist]
top_genres = Counter(genre_list).most_common(10)
genre_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])
genre_df.plot(kind='bar', x='Genre', y='Count', color='skyblue')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Ratings distribution
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index, palette="coolwarm")
plt.title("Distribution of Ratings")
plt.show()
```

```python
#directors with most titles-
top_directors = df['director'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="plasma")
plt.title("Top 10 Directors with Most Netflix Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='black').generate(" ".join(genre_list))
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Netflix Genres Word Cloud")
plt.show()
```
